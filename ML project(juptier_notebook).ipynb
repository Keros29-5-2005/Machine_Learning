```python

import numpy as np
import warnings 
import pandas as pd
import matplotlib.pyplot as mat
warnings.simplefilter("ignore")
%matplotlib inline
data = pd.read_excel("C:\\Users\\solit\\Downloads\\Beer_sales_with_analysis.xlsx")
print(data)
```

```python
new_data=data[['Week','PRICE 18PK']]
```

```python
x = new_data.iloc[:, 0].values.reshape(-1,1)
y= new_data.iloc[:, 1].values.reshape(-1,1)
```

```python
x
y
```

```python
mat.scatter(x,y)
mat.xlable("week")
mat.ylable("price")
mat.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, train_size = 0.8, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python

y_predict = lr.predict(x_test)
```

```python
y_predict 
```

```python
mat.scatter(x_test, y_, color='blue', label='Actual Data')
mat.plot( y_predict ,color='red', label='Regression Line')

mat.xlabel('Week')
mat.ylabel('price')
```
